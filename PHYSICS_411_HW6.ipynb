```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
from IPython.display import HTML
from matplotlib import animation as ani
def RK4(func,h,t,y_begin):
  k1=func(t,y_begin)
  k2=func(t+h/2,h/2*k1+y_begin)
  k3=func(t+h/2,y_begin+h/2*k2)
  k4=func(t+h,y_begin+h*k3)
  return y_begin+h/6*(k1+2*k2+2*k3+k4)
def V_in(t):
  if math.floor(2*t)%2==0:
    return 1
  return -1
RC=0.01
def f_1(t,v_out):
  return 1/RC*(V_in(t)-v_out)
t_1=np.linspace(0,10,1000)
y_1=[0]
for i in range(len(t_1)-1):
  y_1.append(RK4(f_1,0.01,t_1[i],y_1[-1]))
plt.plot(t_1,y_1)
plt.title('RC=0.01')
plt.show()
RC=0.1
y_1=[0]
for i in range(len(t_1)-1):
  y_1.append(RK4(f_1,0.01,t_1[i],y_1[-1]))
plt.plot(t_1,y_1)
plt.title('RC=0.1')
plt.show()
RC=1
y_1=[0]
for i in range(len(t_1)-1):
  y_1.append(RK4(f_1,0.01,t_1[i],y_1[-1]))
plt.plot(t_1,y_1)
plt.title('RC=1')
plt.show()
```

```python
def RK4_p(x_p,y_p,t,h,x_begin,y_begin):
  k1_x=x_p(t,x_begin,y_begin)
  k1_y=y_p(t,x_begin,y_begin)
  k2_x=x_p(t+h/2,x_begin+h/2*k1_x,y_begin+h/2*k1_y)
  k2_y=y_p(t+h/2,x_begin+h/2*k1_x,y_begin+h/2*k1_y)
  k3_x=x_p(t+h/2,x_begin+h/2*k2_x,y_begin+h/2*k2_y)
  k3_y=y_p(t+h/2,x_begin+h/2*k2_x,y_begin+h/2*k2_y)
  k4_x=x_p(t+h,x_begin+h*k3_x,y_begin+h*k3_y)
  k4_y=y_p(t+h,x_begin+h*k3_x,y_begin+h*k3_y)
  return x_begin+h/6*(k1_x+2*k2_x+2*k3_x+k4_x), y_begin+h/6*(k1_y+2*k2_y+2*k3_y+k4_y)
def x_2_p(t,x,y):
  alpha=1
  beta=0.5
  return alpha*x-beta*x*y
def y_2_p(t,x,y):
  gamma=0.5
  delta=2
  return gamma*x*y-delta*y
t_2,h_2=np.linspace(0,30,1000,retstep=True)
x_2=[]
y_2=[]
for t in t_2:
  if t==0:
    p,q=RK4_p(x_2_p,y_2_p,t,h_2,2,2)
    x_2.append(p)
    y_2.append(q)
  else:
    p,q=RK4_p(x_2_p,y_2_p,t,h_2,x_2[-1],y_2[-1])
    x_2.append(p)
    y_2.append(q)
plt.plot(t_2,x_2)
plt.plot(t_2,y_2)
plt.legend(['x(number of rabbits)','y(number of fox)'])
```
